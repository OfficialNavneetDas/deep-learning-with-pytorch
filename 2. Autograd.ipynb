# Autograd

Autograd is a core component of pytorch that provides automatic differentation operations for operations.
it enables gradient computations, which is essential for traning machine learning models using optimization algos like gradient descent

# Basic Steps to Use Autograd

if x = 3 and y=x**2  
so:  
y=9,  
**finding- $\frac{dy}{dx}$ = 2x = 6**

```python
import torch
```

```python
'''the tensor should be in floating numbers'''
x=torch.tensor([3.0],requires_grad=True)
x
```

```python
# make a function
y=x**2
y
```

# backward tracking and finding gradient

```python
# only for scaler inputs
y.backward()
```

```python
x.grad
```

# removing the pre save gradient

```python
x.grad.zero_()
```

# disable gradient tracking

## requires_grad_(False)

```python
x=torch.tensor([3.0],requires_grad=True)
y=x**2
y
```

```python
x.requires_grad_(False)
y=x**2
y
```

## detach()

```python
x=torch.tensor([3.0],requires_grad=True)
y=x**2
y
```

```python
z=x.detach()
y=z**2
y
```

## torch.no_grad()

```python
x=torch.tensor([3.0],requires_grad=True)
y=x**2
y
```

```python
with torch.no_grad():
    y=x**2
y
```
