```python
# tokenize
tokens = word_tokenize(document)
```

```python
# build vocab
vocab={"<UKN>":0}

for token in Counter(tokens).keys():
    if token not in vocab:
        vocab[token]=len(vocab)
```

```python
# extract scentence from data
input_sentence = document.split("\n")
```

```python
def text_to_indices(text,vocab):
    num_text=[]
    for token in text:
        if token in vocab:
            num_text.append(vocab[token])
        else:
            num_text.append(vocab["<UKN>"])
    return num_text
```

```python
input_numerical_sentence=[]
for sentence in input_sentence:
    print(word_tokenize(sentence.lower()))
    print(text_to_indices(word_tokenize(sentence.lower()),vocab))
    input_numerical_sentence.append(text_to_indices(word_tokenize(sentence.lower()),vocab))
```

```python
traning_sequence=[]
for sentence in input_numerical_sentence:
    for i in range(1,len(sentence)):
        traning_sequence.append(sentence[:i+1])
```

```python
x = padded_traning_sequence[:,:-1]
y = padded_traning_sequence[:,-1]
print(x.shape)
print(y.shape)
```

```python
class CustomDataset(Dataset):
    def __init__(self,x,y):
        self.x = x
        self.y = y
    def __len__(self):
        return self.x.shape[0]
    def __getitem__(self,index):
        return self.x[index],self.y[index]
```

```python
dataset = CustomDataset(x,y)
len(dataset)
```

```python
dataset[5]
```

```python
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)
```

# MODEL ARCHITECTURE

```python
class LSTMModel(nn.Module):
    def __init__(self, vocab_size):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size,embedding_dim=100)
        self.LSTM = nn.LSTM(100,150,batch_first=True)
        self.fc = nn.Linear(150,vocab_size)
        
    def forward(self,x):
        embedded = self.embedding(x)
        intermediate_hidden_states, (final_hidden_state, final_cell_state) = self.LSTM(embedded)
        output = self.fc(final_hidden_state.squeeze(0))
        return output
```

# TRANING LOOP

```python
model = LSTMModel(len(vocab))
```

```python
# device = torch.device("cuda" if torch.cuda.is_available() else "cpu") 
# model.to(device)
```

```python
epochs = 50
Learning_rate = 0.0001

loss_function = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(),lr=Learning_rate)
```

```python
# traning loop
for epoch in range(epochs):
    total_loss = 0
    for batch_x, batch_y in dataloader:
        # batch_x, batch_y = batch_x.to(device), batch_y.to(device)
        optimizer.zero_grad()

        output = model(batch_x)
        loss = loss_function(output,batch_y)
        loss.backward()
        optimizer.step()
        total_epoch_loss = total_loss + loss.item()
    # priniting the avrage loss per epoch
    print(f"Epoch: {epoch+1} , total_epoch_loss:{total_epoch_loss:.4f}" )
```

# PREDICTION

```python
def prediction(model, vocab, text):
    # tokenize
    tokenized_text = word_tokenize(text.lower())

    # text to numerical indices
    numerical_indices = text_to_indices(tokenized_text, vocab)

    # padding
    padded_text = torch.tensor([0]*(max_length_word-len(numerical_indices))+numerical_indices,dtype=torch.long)

    # semd to model
    output = model(padded_text)
    # predicted text
    probs =torch.nn.functional.softmax(output)
    
    value, index = torch.max(output, dim=0)

    # merge with the text
    return text+" "+list(vocab.keys())[index]
```
